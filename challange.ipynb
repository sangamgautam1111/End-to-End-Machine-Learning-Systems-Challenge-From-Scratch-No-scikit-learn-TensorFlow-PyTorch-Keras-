# üìå Machine Learning Types & Models (From Scratch Roadmap)

Machine Learning is broadly divided into **three major types** based on the availability of labels and the learning mechanism.

---

## 1Ô∏è‚É£ SUPERVISED LEARNING

### üîπ Definition
Supervised Learning is a learning paradigm where the model is trained on **labeled data**, meaning each input has a corresponding output label.  
The goal is to learn a mapping function from inputs (X) to outputs (Y).

Used heavily in **prediction and classification problems**.

---

### üîπ Types of Supervised Learning

#### ‚úÖ 1. Classification
Output variable is **categorical**.

**Examples**: Spam detection, disease classification, fraud detection.

**Models**:
- Logistic Regression
- K-Nearest Neighbors (KNN)
- Support Vector Machine (SVM)
- Naive Bayes
- Decision Tree Classifier
- Random Forest Classifier
- Gradient Boosting (XGBoost, LightGBM, CatBoost)
- Neural Networks (ANN, CNN)

---

#### ‚úÖ 2. Regression
Output variable is **continuous**.

**Examples**: House price prediction, salary prediction.

**Models**:
- Linear Regression
- Polynomial Regression
- Ridge Regression
- Lasso Regression
- Elastic Net
- Decision Tree Regressor
- Random Forest Regressor
- Gradient Boosting Regressor
- Neural Networks (ANN)

---

### üîπ Key Concepts
- Loss Function
- Bias‚ÄìVariance Tradeoff
- Overfitting & Underfitting
- Train / Test Split
- Evaluation Metrics (Accuracy, Precision, Recall, RMSE)

---

## 2Ô∏è‚É£ UNSUPERVISED LEARNING

### üîπ Definition
Unsupervised Learning deals with **unlabeled data**.  
The model discovers **hidden patterns, structures, or relationships** in data.

Used in **exploration and pattern discovery**.

---

### üîπ Types of Unsupervised Learning

#### ‚úÖ 1. Clustering
Groups similar data points together.

**Examples**: Customer segmentation, image grouping.

**Models**:
- K-Means
- Hierarchical Clustering
- DBSCAN
- Gaussian Mixture Models (GMM)
- Mean Shift

---

#### ‚úÖ 2. Dimensionality Reduction
Reduces number of features while retaining information.

**Examples**: Data visualization, noise reduction.

**Models**:
- Principal Component Analysis (PCA)
- Linear Discriminant Analysis (LDA)
- Singular Value Decomposition (SVD)
- t-SNE
- UMAP
- Autoencoders

---

#### ‚úÖ 3. Association Rule Learning
Finds relationships between variables.

**Examples**: Market basket analysis.

**Models**:
- Apriori Algorithm
- FP-Growth Algorithm
- Eclat

---

### üîπ Key Concepts
- Distance Metrics
- Feature Scaling
- Curse of Dimensionality
- Silhouette Score
- Elbow Method

---

## 3Ô∏è‚É£ REINFORCEMENT LEARNING

### üîπ Definition
Reinforcement Learning is a learning method where an **agent interacts with an environment** and learns by receiving **rewards or penalties**.

Goal is to **maximize cumulative reward** over time.

Used in **decision-making problems**.

---

### üîπ Core Components
- Agent
- Environment
- State
- Action
- Reward
- Policy
- Value Function

---

### üîπ Types of Reinforcement Learning

#### ‚úÖ 1. Value-Based Methods
Learn value of actions or states.

**Models**:
- Q-Learning
- SARSA
- Deep Q Network (DQN)
- Double DQN
- Dueling DQN

---

#### ‚úÖ 2. Policy-Based Methods
Directly learn the policy.

**Models**:
- REINFORCE
- Policy Gradient
- Actor-Critic

---

#### ‚úÖ 3. Model-Based Methods
Learn environment dynamics.

**Models**:
- Markov Decision Process (MDP)
- Monte Carlo Methods
- Dynamic Programming

---

#### ‚úÖ 4. Advanced / Deep Reinforcement Learning (MAANG Level)
- Proximal Policy Optimization (PPO)
- Trust Region Policy Optimization (TRPO)
- Deep Deterministic Policy Gradient (DDPG)
- A3C / A2C
- AlphaZero (Hybrid)

---

### üîπ Real-World Applications
- Robotics
- Game AI
- Recommendation Systems
- Autonomous Driving
- Trading Systems

---

|--------------|------|------|-----------|
